## 1. 介绍一下TCP/IP模型和OSI模型的区别


OSI是国际标准组织(ISO)规定的计算机和通信系统之间进行交流的规范. 一共有7层, 每一次都建立在下面一层的基础上, 并为上面一层提供服务. 由下到上7层分别是: 物理层, 数据链路层, 网络层, 传输层, 会话层, 表示层, 应用层. 虽然很全面, 但是应用更为广泛的是TCP/IP模型.

TCP/IP模型一共有4层, 每层各司其职, 共同实现数据传输.

**应用层:**与OSI模型的应用层、表示层、会话层功能类似, 为用户提供直接的交互接口, 进行网络服务, 如网页浏览(HTTP)

**传输层:**与OSI模型的传输层功能相似, 进行端与端之间的数据传送, 提供可靠数据传送或无连接数据传送, 一般分为两类TCP和UDP. 其中, TCP进行可靠数据传送, UDP进行无连接数据传送(不可靠数据传送)

**网络层:**与OSI模型的网络层功能相似, 将数据进行在路由上传递和转发. 通过路由选择算法选择最优路由路径, 从源主机传递到目的主机, 使用IP地址表示源主机, 进行逻辑地址寻址

**网络接口层:**与OSI模型的物理层和数据链路层功能相似. 进行数据在物理介质上的传输, 例如WIF等, 同时还会进行错误校验和校准, 记录MAC地址.

## 2. 从输入URL到页面展示到底发生了什么?

1. 输入网址, 解析URL, 准备发送HTTP请求

2. 在浏览器缓存中查找是否有该请求, 如果有则直接返回; 反之, 进行下一步

3. DNS解析: 网络请求前, 进行DNS解析, 然后获取要访问的域名的ip地址, 然后根据本地浏览器缓存->本地Host缓存->路由器缓存->DNS服务器缓存->根DNS服务器缓存的顺序进行查找对应的ip地址, 直到找到对应的地址返回返回.(如果是HTTPS, 在解析DNS后, 还要进行TLS链接)

4. TCP三次握手进行链接

5. 创建连接后, 客户端发送请求, 由浏览器端创建请求行、请求头等信息, 并把相关的Cookie等信息包含在内, 然后想目标服务器发送, 如果是HTTPS请求, 还要进行加密

6. 服务器接收到请求, 并进行处理, 然后返回 HTTP头部, 包括状态码, 头部, 内容等, 如果是HTTPS, 还要进行解密.

7. 客户端收到返回响应后进行TCP四次挥手断开连接

8. 浏览器解析响应并渲染:
    * 解析响应头, 如果状态码是301或者302, 则跳转到新地址, 如果是字符串流, 那么给下载管理器处理, 如果是HTML文件, 则进行下一步

    * 解析HTML文件, 生成DOM树, 然后根据CSS文件样式计算, 然后和DOM树一起生成渲染树, 然后进行渲染布局展示.

## 3. HTTP请求报文和响应报文是怎么样的, 有哪些常见的手段?

HTTP报文分为两种, 分别是请求报文和相应报文.

**请求报文**

请求报文由4部分组成, 请求行、请求头、空行、请求体4部分组成.

**请求行:**

* 方法: 如常见的GET方法或者POST方法等
* 资源路径: 请求路径的URL
* HTTP版本

**请求头:**内容有很多种

* Content Length: 请求体的长度
* Content Type: 请求体的类型
* Accpet: 客户端可以接受的数据类型
* Accpert Encoding: 客户端可以解码的内容编码
* Cookie: 存储客户端的cookie数据
* Authorization: 用来进行认证的凭证信息, 如tokens
* Host: 服务器的域名
* ......
* If-None-Match：资源的ETag值，用于缓存控制。
* Connection：管理连接的选项，如 keep-alive。

**空行:** 空行只有一个, 用来分割请求头和请求体

**请求体:** 通常是POST方法或PUT方法的请求体, 包含给服务器的数据

**响应报文**

响应报文和请求报文类似, 有4部分组成, 分别是状态行、响应头、空行、响应体

**状态行:**

包含3个部分, 分别是:HTTP版本、状态码、状态信息, 例如: HTTP/1.1 200 OK

**响应头:**

和请求头一样, 由多种数据组成, 常见如下:
* Content Length: 响应体长度
* Conetent Type: 响应体数据类型
* Server: 服务器信息
* Set Cookie: 添加Cookie信息
* Last Modified: 最后更改资源的时间
* ETag: 响应体的实体标签，用于缓存和条件请求
* ......
* Expires: 响应的过期时间，之后内容被认为是过时的。
* Location：在重定向时指定新的资源位置。
* Access-Control-Allow-Origin: 跨源资源共享（CORS）策略，指示哪些域可以访问资源。

**空行:** 空行只有一个, 用来分割响应头和响应体

**响应体:** 通常是GET方法或者HEAD方法的响应体, 包含返回的数据

## 4.HTTP有哪些请求方式?

* GET: 直接获取指定资源

* POST: 向指定的资源提供数据, 进行数据处理

* PUT: 更新指定的资源

* DELETE: 删除指定资源

* OPTIONS: 查询服务器支持的请求方式

* HEAD: 获取报文首部, 不返回报文主体

* PATCH: 部分更新指定资源

## 5. GET请求和POST请求的区别

1. 用途: GET请求是用来获取数据的, POST请求是用来提交数据的

2. 参数位置: GET请求的参数位置位于URL上, POST请求的参数在请求体上

3. 参数数量、长度: GET请求的参数受URL长度影响, 长度和数量受限; POST请求参数写在请求体上, 理论上不受限制.

4. 安全性: 因为GET请求的参数写在URL上, 安全性较低; POST请求的参数下写在请求体上, 安全性较高

5. 幂等性: GET请求的多次请求遵循幂等性, POST请求不遵循, 因为每次POST请求都有可能更改资源.

6. 缓存: GET请求可以被缓存, POST请求默认不会被缓存.

## 7. 什么是强缓存和协商缓存

强缓存:

定义: 客户端在没有访问服务器的情况下, 直接访问本地缓存获取所需资源.

方式: 

Expires缓存: 这是一个强缓存时间段, 在这个范围内可以使用强缓存返回资源. 但是会因为因为本地时间不准而失败, 已经被弃用.

Cache-Control: 相比Expires, Cache-Control 是一个相对时间段, 相对于服务器的当前时间. 例如, Cache-Control: max-age=3600, 表明在3600秒内, 客户端可以使用强缓存返回资源.

协商缓存:

定义: 在强缓存失败之后, 客户端会向服务器发送包含Last-Modified或ETag信息的请求, 如果服务器通过对比发来的Last-Modified信息和自己的Last-Modified信息是一样的或者ETag哈希值是一样的(资源没有变化), 则返回304 Not Modified 信息, 客户端就可以直接读取本地缓存进行资源访问; 反之, 如果发生了变化, 则返回跟新之后的资源.

Last-Modified协商缓存过程:

客户端在第一次请求时, 服务器返回 Last-Modified 头部, 告诉客户端上次返回的资源最后修改时间; 客户端在下次请求时, 客户端在请求头部添加 If-Modified-Since 头部, 并将上次返回的 Last-Modified 值作为值.

Last-Modified的缺点:

Last-Modified 信息是以秒为单位, 如果在一秒内修改了资源, 则无法区分出资源是否发生了变化, 进而无法返回最新资源; 同时, 如果发生了修改后又改回去的现象, 这依旧属于新资源, 会造成时间的浪费.

ETag协商缓存过程:

客户端在第一次请求时, 服务器返回 ETag 头部, 告诉客户端上次返回的资源 ETag 值; 客户端在下次请求时, 客户端在请求头部添加 If-None-Match 头部, 并将上次返回的 ETag 值作为值.

ETag的优点: 

将原先协商缓存的比较时间戳的形式修改成了比较文件指纹（根据文件内容计算出的唯一哈希值）。

ETag 值是根据资源的内容生成的, 所以它在一秒内发生了变化, 也能区分出资源是否发生了变化. 并且, ETag 值可以是任意值, 完全可以是 MD5, SHA-1, 甚至是 UUID.

ETag 值可以让客户端在下次请求时, 直接通过 If-None-Match 头部来告诉服务器上一次返回的 ETag 值, 让服务器判断该资源是否有变化.

官方解析:

强缓存和协商缓存是HTTP缓存机制的两种类型，它们用于减少服务器的负担和提高网页加载速度。


1. 强缓存：客户端在没有向服务器发送请求的情况下，直接从本地缓存中获取资。
* Expires强缓存：设置一个强缓存时间，此时间范围内，从内存中读取缓存并返回。但是因为Expires判断强缓存过期的机制是获取本地时间戳，与之前拿到的资源文件中的Expires字段的时间做比较来判断是否需要对服务器发起请求。这里有一个巨大的漏洞：“如果我本地时间不准咋办？”所以目前已经被废弃了。
* Cache-Control强缓存：目前使用的强缓存是通过HTTP响应头中的Cache-Control字段实现，通过max-age来告诉浏览器在指定时间内可以直接使用缓存数据，无需再次请求。

2. 协商缓存：当强缓存失效时，浏览器会发送请求到服务器，通过ETag或Last-Modified等HTTP响应头与服务器进行验证，以确定资源是否被修改。如果资源未修改，服务器返回304 Not Modified状态码，告知浏览器使用本地缓存；如果资源已修改，则返回新的资源，浏览器更新本地缓存。这种方式需要与服务器通信，但可以确保用户总是获取最新的内容。

* 基于Last-Modified的协商缓存

  * Last-Modified 是资源的最后修改时间，服务器在响应头部中返回。
  * 当客户端读取到Last-modified的时候，会在下次的请求标头中携带一个字段:If-Modified-Since，而这个请求头中的If-Modified-Since就是服务器第一次修改时候给他的时间
  * 服务器比较请求中的 If-Modified-Since 值与当前资源的 Last-Modified 值，如果比对的结果是没有变化，表示资源未发生变化，返回状态码 304 Not Modified。如果比对的结果说资源已经更新了，就会给浏览器正常返回资源，返回200状态。
* 但是这样的协商缓存有两个缺点：
  * 因为是更改文件修改时间来判断的，所以在文件内容本身不修改的情况下，依然有可能更新文件修改时间（比如修改文件名再改回来），这样，就有可能文件内容明明没有修改，但是缓存依然失效了。
  * 当文件在极短时间内完成修改的时候（比如几百毫秒）。因为文件修改时间记录的最小单位是秒，所以，如果文件在几百毫秒内完成修改的话，文件修改时间不会改变，这样，即使文件内容修改了，依然不会返回新的文件。

* 基于ETag的协商缓存：将原先协商缓存的比较时间戳的形式修改成了比较文件指纹（根据文件内容计算出的唯一哈希值）。

  * ETag 是服务器为资源生成的唯一标识符（文件指纹），可以是根据文件内容计算出的哈希值，服务端将其和资源一起放回给客户端。
  * 客户端在请求头部的 If-None-Match 字段中携带上次响应的 ETag 值。
  * 服务器比较请求中的 If-None-Match 值与当前资源的 ETag 值，如果匹配，表示资源未发生变化，返回状态码 304 Not Modified。如果两个文件指纹不吻合，则说明文件被更改，那么将新的文件指纹重新存储到响应头的ETag中并返回给客户端

## 8. HTTP1.0和HTTP1.1的区别

持久连接:

HTTP1.1默认打开持久连接, 允许一个TCP上进行多次请求和响应, 减少断开和连接所需要的时间; HTTP1.0则不是, 它的持久连接需要Connect字段的keep-alive来指明, 它默认的是短连接, 一个TCP链接只能处理一次请求和响应.

缓存处理方面:

HTTP1.1使用if-None-Match和ETag进行缓存处理; HTTP1.0使用if-Modified-Since和Last-Modified进行缓存处理.

管道通信:

HTTP1.1支持管道通信(非默认打开), 可以在第一个响应没到之前发送多个请求; HTTP1.0不支持

错误处理:

HTTP1.1添加了新的报错处理状态行, 例如100 continue, 用于增强错误处理和请求的中间响应

宽带处理:

HTTP1.0在接受一次GET请求后会将所有的资源都返回; 而HTTP1.1则会将所需要的部分资源返回, 也就是206状态码

Host头:

HTTP1.1允许客户端指定所要请求的主机名, 这样一个服务器上就能代理多个域名了; HTTP1.0没有.

## 10. HTTP3.0有了解过吗?

HTTP3.0是HTTP的最新版本, 使用QUIC协议, 有以下特性:

1. 无队头阻塞

HTTP3.0使用QUIC协议, QUIC协议是基于UDP的协议, 所以多个Steam之间无需连接, 可以直接传送数据, 即使有一个Steam掉包了, 也不影响其他Steam的数据传送

2. 安全问题

HTTP3.0使用TLS加密, 安全性更高

3. 无RTT连接建立

HTTP3.0支持在第一次握手后, 进行零RTT链接, 减少了连接延迟, 可以更快加载资源.

4. 向前纠错机制

HTTP3.0在发送一个数据包的时候, 还会存放上一个数据包的部分内容, 这样即使上一个数据出现部分数据丢失现象, 也可以凭借后一个数据包中的冗余数据进行补充和检验, 减少重新发送的时间消耗. 虽然这样的数据传送会对当前数据包的有数量限制, 但是减少了重发的次数

5. 连接迁移

HTTP3.0允许在切换网络时(如: wife改成了流量), 保存现在的链接, 节省了断连、重连、重缓存的时间消耗

## 11. HTTP和HTTPS的区别

1. HTTP的端口号是80, HTTPS的端口号是443

2. HTTP连接只需要三次握手即可, 而HTTPS还需要进行SSL/TLS连接

3. HTTPS相较于HTTP, 又添加了SSL/TLS保密协议, 更加安全

4. HTTP协议需要向CA申请数字证书, 来保证服务器的身份是可信的

## 12. HTTPS的工作原理(HTTPS创建连接的过程)

1. 客户端发送HTTPS请求后, 服务器会将其公钥返回给客户端

2. 客户端检验公钥是不是CA认可的, 检验公钥的安全性和可靠性

3. 验证无误后, 随机生成一个对称加密密钥, 然后用服务器给的公钥进行加密这个客户端的对称加密密钥, 然后返回给服务器

4. 服务器接受到后, 根据自己的私钥进行解密得到客户端的对称加密密钥, 这样客户端和服务端就有相同的对称加密密钥了, 可以进行加密和解密了

5. 客户端将想要发送的数据, 用对称加密密钥进行加密, 然后发送给服务端, 服务端收到之后, 用对称加密密钥进行解密, 返回响应时, 也是用对称加密密钥加密, 在客户端进行解密, 以此实现了信息的加密传输, 同时这也是零RTT可靠的主要原因.

6. SSL/TLS协议还包括消息完整校验机制, 如消息验证码, 确保数据在传输过程中没有被篡改.

7. 数据传输完成后, 客户端和服务端的密钥会被销毁, 确保不会留下安全隐患

## 13. TCP和UDP的区别

1. TCP提供连接的数据传送, 提供可靠的数据传送, UDP不提供, UDP是无连接传输

2. TCP有拥塞控制机制, 会根据网络情况调整传输效率, UDP没有, 它是稳定的数据传输效率

3. TCP通过滑动窗口控制, 调整流量情况, 避免发送数据包够多, 导致接收方处理不过来, UDP没有

4. TCP的报文头部比较复杂, 包含序列号、确认号等信息用于验证, UDP报文头部比较简单

5. TCP服务有检查和重传功能, 对于丢失的或者不完整的数据包可以检测处理, 然后重传; UDP不能

6. 由于TCP的连接建立、数据检验和重传, 会占有更多的网络资源, 开销更大

7. 使用场景:

TCP: 网页浏览、文件传输

UDP: 实时通话、语音通话

## 14. TCP连接是如何确保可靠性的

TCP通过差错控制(序列号、确认应答、数据校验), 超时重传, 拥塞控制, 流量控制来确保:

**序列号:**

每次请求都会包含一个序列号, 用来标识该数据包, 用来确保数据包顺序正确.

**确认应答:**

服务器在接收到数据包并处理完成后, 会将包含序列号的响应报文返回给客户端, 如果超时没有确认, 这重传

**数据校验:**

用来检验数据是否完整正确, 如果数据包有缺失, 则丢掉该数据包, 等待客户端确认应答超时之后, 再次发送.

**超时重传:**

当客户端在一段时间内没有接受到来自服务端的确认应答, 则会重新发送超时的数据包.

**流量控制:**

TCP会使用滑动窗口控制发送的数据包个数, 确保可以更加充分的使用服务端的处理请求的数量, 如果服务器处理不过来, 则滑动窗口数减少; 如果利用不充分, 则增加滑动窗口数.

**拥塞控制:**

TCP通过慢启动、快速重启、拥塞控制等, 来控制数据的发送数量(滑动窗口数), 避免网络阻塞.


## 15. 既然提到了拥塞控制，那你能说说说拥塞控制是怎么实现的嘛

**慢启动:**

在开始发送数据包时, 先从一次发送1个数据包开始, 然后呈指数形增加, 逐步增加发送窗口的数量, 以此避免因一开始一次性发送过多数据包而产生网络拥塞.

**拥塞避免:**

当慢启动发送的数据包个数到达一定阈值的时候, 会进入线性增长阶段, 这一阶段被称为拥塞避免, 避免因为指数增长过快, 而直接出现网络拥塞的情况.

**快速重传:**

当客户端收到三次确认相同的ACK时, 直接确定出现数据包丢失现象, 不必等待超时, 直接重传丢失的数据包.

**快速恢复:**

当出现快速重传时, 会进入快速恢复阶段, 将当前的拥塞窗口数设置为出现快速重传时的窗口数/2 + 已确认但未被快速重传的数据块的数量 的数量, 然后阈值设置为出现快速重传时的窗口数/2,进入拥塞避免阶段.

## 16. TCP流量控制是怎么实现的？

* 发送方TCP的报文中有一个用来存放可以接受的滑动窗口数信息的位置, 发送给接收端

* 接收端获取到报文后, 会根据自身的缓存窗口数, 在存放滑动窗口数的位置填写相关数据, 然后返回给客发送端

* 发送端接受到之后, 在再根据自己的发送缓存区的数量和网络拥塞窗口数量, 选择其中最小的作为滑动窗口数量.

**目的:**

为了不让请求数量超多接收方的缓存区而丢失, 确保接收方来得及处理发送的信息.

**确认机制**

接收方会定期发送确认（ACK）报文，告知发送方已成功接收数据。这也与流量控制密切相关，因为接收方可以通过ACK报文中的窗口字段来通知发送方它的当前窗口大小。(算是和拥塞控制的联系)

## 17. UDP怎么实现可靠传输

UDP它不属于连接型协议，因而具有资源消耗小，处理速度快的优点，所以通常音频、视频和普通数据在传送时使用UDP较多，因为它们即使偶尔丢失一两个数据包，也不会对接收结果产生太大影响。传输层无法保证数据的可靠传输，只能通过应用层来实现了。实现的方式可以参照tcp可靠性传输的方式，只是实现不在传输层，实现转移到了应用层。关键在于两点，从应用层角度考虑：

（1）提供超时重传，能避免数据报丢失。

（2）提供确认序列号，可以对数据报进行确认和排序。

**本端：**首先在UDP数据报定义一个首部，首部包含确认序列号和时间戳，时间戳是用来计算RTT(数据报传输的往返时间)，计算出合适的RTO(重传的超时时间)。然后以等-停的方式发送数据报，即收到对端的确认之后才发送下一个的数据报。当时间超时，本端重传数据报，同时RTO扩大为原来的两倍，重新开始计时。

**对端：**接受到一个数据报之后取下该数据报首部的时间戳和确认序列号，并添加本端的确认数据报首部之后发送给本端。根据此序列号对已收到的数据报进行排序并丢弃重复的数据报。

## 18. TCP连接三次握手的过程，为什么是三次，可以是两次或者更多吗？

(1) 三次握手的过程

**第一次握手：**客户端向服务器发送一个SYN （同步序列编号）报文，请求建立连接，客户端进入SYN_SENT 状态。

**第二次握手：**服务器收到SYN 报文后，如果同意建立连接，则会发送一个SYN-ACK （同步确认）报文作为响应，同时进入SYN_RCVD 状态。

**第三次握手：**客户端收到服务器的SYN-ACK 报文后，会发送一个ACK （确认）报文作为最终响应，之后客户端和服务器都进入ESTABLISHED 状态，连接建立成功。


(2)为什么需要三次握手

通过三次握手，客户端和服务器都能够确认对方的接收和发送能力。第一次握手确认了客户端到服务器的通道是开放的；第二次握手确认了服务器到客户端的通道是开放的；第三次握手则确认了客户端接收到服务器的确认，从而确保了双方的通道都是可用的。

而如果仅使用两次握手，服务器可能无法确定客户端的接收能力是否正常，比如客户端可能已经关闭了连接，但之前发送的连接请求报文在网络上延迟到达了服务器，服务器就会主动去建立一个连接，但是客户端接收不到，导致资源的浪费。而四次握手可以优化为三次。

## 19. TCP连接四次挥手的过程，为什么是四次？

（1）四次挥手的过程

**第一次挥手：**客户端发送一个FIN报文给服务端，表示自己要断开数据传送，报文中会指定一个序列号 (seq=x)。然后,客户端进入FIN-WAIT-1 状态。

**第二次挥手：**服务端收到FIN报文后，回复ACK报文给客户端，且把客户端的序列号值+1，作为ACK报文的序列号(seq=x+1)。然后，服务端进入CLOSE-WAIT(seq=x+1)状态，客户端进入FIN-WAIT-2状态。

**第三次挥手：**服务端也要断开连接时，发送FIN报文给客户端，且指定一个序列号(seq=y+1)，随后服务端进入LAST-ACK状态。

**第四次挥手：**客户端收到FIN报文后，发出ACK报文进行应答，并把服务端的序列号值+1作为ACK报文序列号(seq=y+2)。此时客户端进入TIME-WAIT状态。服务端在收到客户端的ACK 报文后进入CLOSE 状态。如果客户端等待2MSL没有收到回复，才关闭连接。

（2）为什么需要四次挥手

TCP是全双工通信，可以双向传输数据。任何一方都可以在数据传送结束后发出连接释放的通知，待对方确认后进入半关闭状态。 当另一方也没有数据再发送的时候，则发出连接释放通知，对方确认后才会完全关闭 TCP 连接。因此两次挥手可以释放一端到另一端的TCP连接，完全释放连接一共需要四次挥手。

只有通过四次挥手，才可以确保双方都能接收到对方的最后一个数据段的确认，主动关闭方在发送完最后一个ACK后进入TIME-WAIT 状态，这是为了确保被动关闭方接收到最终的ACK ，如果被动关闭方没有接收到，它可以重发FIN 报文，主动关闭方可以再次发送ACK 。

而如果使用三次挥手，被动关闭方可能在发送最后一个数据段后立即关闭连接，而主动关闭方可能还没有接收到这个数据段的确认。

## 20. HTTP的Keep-Alive是什么？TCP 的 Keepalive 和 HTTP 的 Keep-Alive 是一个东西吗？

1. HTTP 的 Keep-Alive，是由应用层实现的，称为 HTTP 长连接
每次请求都要经历这样的过程：建立 TCP连接 -> HTTP请求资源 -> 响应资源 -> 释放连接，这就是HTTP短连接，但是这样每次建立连接都只能请求一次资源，所以HTTP 的 Keep-Alive实现了使用同一个 TCP 连接来发送和接收多个 HTTP 请求/应答，避免了连接建立和释放的开销，就是 **HTTP 长连接**。通过设置HTTP头Connection: keep-alive来实现。

2. TCP 的 Keepalive，是由TCP 层（内核态）实现的，称为 TCP 保活机制，是一种用于在 TCP 连接上检测空闲连接状态的机制

当TCP连接建立后，如果一段时间内没有任何数据传输，TCP Keepalive会发送探测包来检查连接是否仍然有效。
