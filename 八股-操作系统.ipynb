# 八股文

## 1. 进程和线程之间有什么区别

**进程**是资源分配和调度基本单位

**线程**是程序运行的基本单位, 是**进程**子任务, 一个进程至少有一个线程, 当一个进程有多个线程时, 这些线程占用同一块内存.

资源开销方面

进程: 每个进程都有自己独立的空间, 进程的创建和销毁开销比较大; 进程的切换需要切换和恢复整个进程上下文资源, 进程切换开销大

线程: 在一个进程中的线程共同占有一块空间, 线程的创建和销毁开销比较小; 线程的切换只需要切换和恢复部分线程上下文资源, 线程切换开销小

通信方面

进程: 因为进程拥有独立的空间, 所以进程之间的通信需要消息队列、管道通信等技术

线程: 在同一个进程内的线程, 因为他们共享相同的内存, 所以线程之间的通信更简单

安全性方面

进程: 由于每个进程拥有自己独立的空间, 所以, 当一个进程出现问题时, 不会影响到其他进程, 安全性较高

线程: 在同一个进程内的线程, 因为他们共享相同的内存, 所以, 当一个线程出现问题时, 可能会影响到其他线程, 安全性较低
